```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
list_dh = [5, 8, 19 , 39, 79, 148, 247, 391, 611, 922]
list_orca = [20, 37, 78, 146, 262, 408, 650, 954, 1506, 2081]
list_dlpno = [27, 69, 176, 375, 693, 1154, 1736, 2624, 3615, 4941]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(7, 3))
for ax in axs:
    ax.plot(list_dh, label="dh", marker="o")
    ax.plot(list_orca, label="ORCA", marker="o")
    ax.plot(list_dlpno, label="ORCA with DLPNO", marker="o")
    ax.set_xticks(range(10))
    ax.set_xticklabels(range(1, 11))
    ax.legend()
    ax.set_xlabel("C$_n$H$_{2n+2}$")
axs[1].set_yscale("log")
axs[0].set_ylabel("wall time / sec")
axs[0].set_title("linear scale")
axs[1].set_title("logarithmic scale")
fig.tight_layout()
fig.savefig("timing-polar-chain.pdf", transparent=True)
```
