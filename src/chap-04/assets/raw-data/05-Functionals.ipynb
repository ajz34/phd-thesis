```python
def get_libxc_functional(xc):
    try:
        return pylibxc.LibXCFunctional(xc, 1)
    except:
        for prefix in ["LDA_XC_", "LDA_GGA_XC_", "GGA_XC_", "HYB_GGA_XC_", "MGGA_XC_", "HYB_MGGA_XC_"]:
            return get_libxc_functional(prefix + xc, 1)
    assert False, xc
```

```python
def get_doi_low(xc):
    try:
        f = pylibxc.LibXCFunctional(xc, 1)
        return f.get_doi()
    except:
        ps = libxc.parse_xc(xc)
        l = []
        for p in ps[1]:
            f = pylibxc.LibXCFunctional(p[0], 1)
            l += f.get_doi()
        return l
    assert False, xc
```

```python
def get_detail_low(xc):
    is_hyb = is_rsh = False
    xc_type = None
    ni = dft.numint.NumInt()
    omega, a, b = ni.rsh_and_hybrid_coeff(xc)
    if omega > 0:
        is_rsh = is_hyb = True
    elif a != 0 or b != 0:
        is_hyb = True
    xc_type = ni.libxc.xc_type(xc)
    return {"type": xc_type, "hybrid": is_hyb, "range-separate": is_rsh}
```
